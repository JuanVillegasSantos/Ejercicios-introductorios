```python
#Ley de movimiento planetario: T=Cx^(3/2); T=Periodo[dias]; 
#x=Distancia al sol [Mkm]; C=kte
#Datos Medidos:
#Mercurio (58,88); Venus (108,225); Tierra (150,365); Marte (228,687)
#Realizar el ajuste de la curva T vs x
#Calcular periodo de Jupiter a 778 Mkm del sol
```

```python
#Imports
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
#Kepler function
def kepler(x,C):
    T=C*x**(3/2)
    return T
#Data
x=np.array([58,108,150,228])#Distance
T=np.array([88,225,365,687])#Period
#Ajuste
P_opt,P_cov=curve_fit(kepler,x,T) #curve_fit usa la aproximaci√≥n de m√≠nimos cuadrados no lineales para ajustar funcion a datos Param (modelo,x,f(x)); return: P_opt=Par√°metro(s) o kte(s)de ajuste; P_cov=Matriz de varianza entre los datos
print('C= ',P_opt[0])
#Plot exp. data
plt.plot(x,T,'*b',label='Exp. Data')
#Plot Modelo ajustado
x_adjst=np.linspace(x[0],x[-1])
T_adjst=kepler(x_adjst,P_opt)
plt.plot(x_adjst,T_adjst,'k',label='fit: C=%.3f' % (P_opt))

plt.xlabel('Distancia al sol [Mkm]')
plt.ylabel('Periodo[dias]')
plt.legend()
#Calculo Jupiter
x_jup=778
T_jup=kepler(x_jup,P_opt)
print('Periodo Jupiter= %.2f d√≠as, calculado con C=%.3f' % (T_jup,P_opt[0]))
```
