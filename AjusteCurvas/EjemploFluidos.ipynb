```python
#Tasa de corte de fluidos: tau=f(gamma)=mu*gamma;tau=esfuerzo corte[Pa]; 
#gamma=Tasa de cizalla[1/s];mu=Viscosidad[Pa-s]
#Datos Medidos:
#[[50,6.01],[70,7.48],[90,8.59],[110,9.19],[130,10.21]]
#Realizar el ajuste de la curva tau vs gamma con minimos cudrados y con ley de potencias
```

```python
#Imports
import numpy as np
from scipy.optimize import curve_fit
from matplotlib import pyplot as plt
#Model
def func(x,mu,n):
    return mu*x**n
#Data
gamma=np.array([50,70,90,110,130])
tau=np.array([6.01,7.48,8.59,9.19,10.21])
plt.plot(gamma,tau,'*k',label='Exp. Data')
#Ajuste y plot minimos cuadrados lineal
p,cov=np.polyfit(gamma,tau,1)#Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y). Returns a vector of coefficients p that minimises the squared error.
t=np.linspace(gamma[0],gamma[-1])
tauAdjst_M=np.polyval([p,cov],t)

plt.plot(t,tauAdjst_M,'g',label='fit')

plt.xlabel('$\dot{\gamma}$ [1/s]')
plt.ylabel('$\\tau$ [Pa]')
plt.legend()
```

```python
#Ajuste y plot ley de potencias
popt, pcov=curve_fit(func, gamma, tau)
tauAdjst_P=func(x,*popt)

plt.plot(gamma,tau,'*k',label='Exp. Data')
plt.plot(t,tauAdjst_P, 'r-',
         label='fit: $\mu$=%5.3f, n=%5.3f' % tuple(popt))
plt.xlabel('$\dot{\gamma}$')
plt.ylabel('$\tau$')
plt.legend()
```
